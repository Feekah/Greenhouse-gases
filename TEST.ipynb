```python
GHC = pd.read_csv('atmosphericGHC.csv')
carbon = pd.read_csv('Carbon.csv')
HFC = pd.read_csv('HydroFC.csv')
methane = pd.read_csv('Methane.csv')
NF3 = pd.read_csv('NitroTriFluoride.csv')
NitrousO = pd.read_csv('NitrousOxide.csv')
PerFC = pd.read_csv('PerFC.csv')
SulHexF = pd.read_csv('SulphurHexFlouride.csv')
```

```python
#Let's start by filtering out row 0 to 27
GHG_per_Industry = GHC[0:28]

#Let's delete the first and second columns since there is no relevant data in them
GHG_per_Industr = GHG_per_Industry.drop(['Total greenhouse gas1 emissions by industry section and group, 1990 to 2020', 'Unnamed: 1'], axis=1)

#Let us also drop the first two and last two rows
GHG_per_Industry = GHG_per_Industr.drop([0,1,24, 26,27])

#Let's reset the distorted index to make things easy:
GHG_per_Industry = GHG_per_Industry.reset_index()

#We aim to make the year rows be the header of this table so let's do it 
GHG_per_Industry = GHG_per_Industry.rename(columns=GHG_per_Industry.iloc[0]).loc[1:]

#Let's remove the second column
GHG_per_Industry = GHG_per_Industry.drop([2], axis=1)

# return the transpose
GHG_per_Industry = GHG_per_Industry.transpose()

GHG_per_Industry.index = range(0, len(GHG_per_Industry))

#Let's change the header to the first row and delete the first row
GHG_per_Industry = GHG_per_Industry.rename(columns=GHG_per_Industry.iloc[0]).loc[1:]

GHG_per_Industry.insert(0, "Year", [1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020], True)

GHG_per_Industry = GHG_per_Industry.reset_index(drop=True)

GHG_per_Industry
```

VISUALIZATIONS
#Let's use plotly to visualize the data

```python
#Let's import the libraries we need for plotly
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

```

```python
#Let's do a subplots for the 6 gases
fig = make_subplots(rows=2, cols=4, subplot_titles=("Carbon Dioxide", "Methane", "Nitrous Oxide", "Perfluorocarbons", "Sulphur Hexaflouride", "Hydrofluorocarbons"))
fig.add_trace(go.Scatter(x=carbon['Year'], y=carbon["Total CO2 emissions"], name="Total CO2 emissions"), row=1, col=1)
fig.add_trace(go.Scatter(x=GHG_per_Industry['Year'], y=GHG_per_Industry["Total greenhouse gas emissions"], name="Total Greenhouse Emissions"), row=1, col=2)
fig.add_trace(go.Scatter(x=methane['Year'], y=methane["Total CH4 emissions"], name="Total (CH4) Methane emissions"), row=1, col=3)
fig.add_trace(go.Scatter(x=NF3['Year'], y=NF3["Total NF3 emissions"], name="Total NF3 emissions"), row=1, col=4)
fig.add_trace(go.Scatter(x=NitrousO['Year'], y=NitrousO["Total N2O emissions"], name="Total Nitrous Oxide emissions"), row=2, col=1)
fig.add_trace(go.Scatter(x=HFC['Year'], y=HFC["Total HFC emissions"], name="Total Hydrofluorocarbon emissions"), row=2, col=2)
fig.add_trace(go.Scatter(x=PerFC['Year'], y=PerFC["Total PFC emissions"], name="Total Perfluorocarbons emissions"), row=2, col=3)
fig.add_trace(go.Scatter(x=SulHexF['Year'], y=SulHexF["Total SF6 emissions"], name="SulHexF"), row=2, col=4)

fig.update_layout(height=600, width=800, title_text="Total Emissions of Greenhouse Gases in UK from 1990 to 2020")
fig.show()

```
